```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
lead = gpd.read_file('processed-spatial/lead/lead_by_state.json')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(20, 14))

lead.plot(column='%>5ug/dl',
          ax=ax,
          legend=True,
          legend_kwds={'label': "Ratio of children with more than 5ug/dl in blood",
                        'orientation': "horizontal"});
```

```python
opio = gpd.read_file('processed-spatial/drugs/opioids_deaths_by_state.json')
opio.head()
```
