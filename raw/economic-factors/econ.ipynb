```python
import pandas as pd
import numpy as np
import geopandas as gpd
```

```python
# we are interested in jobs and income for now
```

```python
jobs = pd.read_csv('Release23_June2021/Jobs.csv')
income = pd.read_csv('Release23_June2021/Income.csv')
```

```python
income.head()
```

```python
# We will use ALL columns
```

```python
states_and_counties = gpd.read_file('../../processed-spatial/general/counties_and_states.json')
```

```python
states_and_counties.head()
```

```python
mini_set = states_and_counties[['STATEFP', 'NAME', 'STUSPS', 'geometry']]
```

```python
# Merge

cjobs = mini_set.merge(jobs, left_on=['STUSPS', 'NAME'], right_on=['State', 'County'], how='left')
```

```python
cjobs.head()
```

```python
cjobs.to_file('counties_jobs.json', driver='GeoJSON')
```

```python
# The same for income
```

```python
# Merge

cinc = mini_set.merge(income, left_on=['STUSPS', 'NAME'], right_on=['State', 'County'], how='left')
```

```python
cinc.to_file('counties_income.json', driver='GeoJSON')
```

```python
# Now agg at a state level
```

```python
mini_set.head()
```

```python
micro_set = mini_set[['STUSPS', 'NAME']]
```

```python
micro_jobs = micro_set.merge(jobs, left_on=['STUSPS', 'NAME'], right_on=['State', 'County'], how='left')
```

```python
states = gpd.read_file('../spatial/us-states/cb_2018_us_state_500k.shp')
```

```python
states.head()
```

```python
mini_states = states[['STUSPS', 'NAME', 'geometry']]
```

```python
micro_income = micro_set.merge(income, left_on=['STUSPS', 'NAME'], right_on=['State', 'County'], how='left')
```

```python
#### AGG
```

```python
micro_jobs.head()
```

```python
agg_jobs = micro_jobs.groupby(['STUSPS']).mean()
```

```python
agg_jobs.head()
```

```python
agg_jobs.drop('FIPS', inplace=True, axis=1)
```

```python
agg_jobs.dropna(how='all', inplace=True)
```

```python
agg_income = micro_income.groupby(['STUSPS']).mean()
agg_income.drop('FIPS', inplace=True, axis=1)
agg_income.dropna(how='all', inplace=True)
```

```python
# Merge both with states
```

```python
agg_income.reset_index(inplace=True)
agg_jobs.reset_index(inplace=True)
```

```python
merged_1 = mini_states.merge(agg_jobs, on='STUSPS')
```

```python
merged = merged_1.merge(agg_income, on='STUSPS')
```

```python
merged.head()
```

```python
merged.to_file('income_jobs_per_state.json', driver='GeoJSON')
```
